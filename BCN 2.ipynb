```python
 y= data2['lifelong calculation']
```

```python
not_satisfied_client=[]
satisfied_client=[]
loyal_client=[]
for i in y:
    if i <499:
       not_satisfied_client.append(1)
    elif 500<i<999:
       satisfied_client.append(1)
    elif i>1000:
        loyal_client.append(1)

```

```python
plt.hist(not_satisfied_client, label="not satisfied clients")
plt.hist(satisfied_client, label="satisfied clients")
plt.hist(loyal_client, label="loyal clients")
plt.title('clients classifications',fontsize=16)
plt.ylabel('No. of clients')
plt.xlabel('type of client')
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(not_satisfied_client, alpha=0.5, label="less than 500 days loyalty")
plt.hist(satisfied_client, alpha=0.5, label="between 500 to 1000 days loyalty")
plt.hist(loyal_client, alpha=0.5, label="more than 1000 days loyalty")
plt.xlabel("type of client", size=14)
plt.ylabel("No. of clients", size=14)
plt.title("clients classifications")
plt.legend(loc='upper right')
#plt.savefig("overlapping_histograms_with_matplotlib_Python_2.png")
```
